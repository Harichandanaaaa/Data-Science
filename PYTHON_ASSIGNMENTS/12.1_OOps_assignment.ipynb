#Question 1: Bank Account System
Create a class BankAccount with the following:

Attributes:
name (account holder name)
balance (initial balance)

Methods:
deposit(amount)
Adds money to balance

withdraw(amount)
Subtracts money if balance is sufficient, otherwise print "Insufficient balance"

display_balance()
Prints current balance

Requirements:
Maintain a transaction history using a list
Use a loop to allow the user to perform multiple operations (deposit, withdraw, exit)
Validate input (amount should be positive)

```python
class BankAccount:
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
        self.history=[]
             
    # Since amount is greated than the 0 only than the amount will be deposited to the balanc.
    ## Deposite is a method()     
    def deposit(self,amount):
        if amount > 0:
            self.balance +=amount
            self.history.append(amount)
            print("Amount deposited succesfully")  
        else:
            print("Enter Positive amount")
            
    # if amount is less than equal to 0 than reject
    #if amount is greater than balance than withdraw
    ##withdraw is a method()       
    def withdraw(self,amount):
        if amount <= 0:
            print("Enter the Positive number")
        elif amount > self.balance:
            print("Insufficient Balnce")   
        else:
            self.balance -= amount
            self.history.append(-amount)
            print("Amount withdrawn successfully")
            
    def display_balance(self):
        print(f"{self.name}'s Current Balance:{self.balance}") 
        
    def show_history(self):
        print("Transaction History: ")   
        if not self.history:
            print("No transaction yet")
        else:
            for item in self.history:
                if item > 0:
                    print(f"Deposited:{item}")
                else:
                    print(f"withdrawn:{-item}")       
                        
## Main coding part
name=input("Enter the name: ")
balance=float(input("Enter initial balance: ")) 
account=BankAccount(name,balance)

while True:
    print("\n1.Deposit")
    print("2.Withdraw")
    print("3.Display Balance")
    print("4.Transaction History")
    print("5.Exit")
    
    choice=input("Enter your choice: ")
    
    if choice=="1":
        amount=float(input("Enter amount: "))
        account.deposit(amount)
    elif choice=="2":
        amount=float(input("Enter amount: "))
        account.withdraw(amount)
    elif choice=="3":
        account.display_balance()
    elif choice=="4":
        account.show_history()  
    elif choice=="5":
        print("Thank you")
        break
    else:
        print("Invalid choice")
                           
    
```

##Question 2: Student Management System
Create a class Student with:

Attributes:
name
marks (dictionary with subject as key and marks as value)

Methods:
calculate_total()
Returns total marks

calculate_average()
Returns average marks

get_grade()

average >= 90 → Grade A

average >= 75 → Grade B

average >= 50 → Grade C

else → Fail

Requirements:
Store multiple student objects in a list

Use a loop to display each student’s name, total, average, and grade

Handle empty marks dictionary case

```python
class Student:
    def __init__(self,name,marks):
       self.name=name
       self.marks=marks
      
    def calculate_total(self):
        return sum(self.marks.values())
    
    def calculate_avg(self):
        if not self.marks:
            return 0
        return self.calculate_total()/len(self.marks)
          
    def get_grade(self):
        avg=self.calculate_avg()
        
        if avg >= 90:
            return "Grade A"      
        elif avg >= 75:
            return "Garde B"
        elif avg >= 50:
            return "Garde C"
        else:
            return "Fail"
            
## main code
students=[]
num=int(input("\nEnter number of students: "))

for i in range(num):
    name=input("\nEnter student name:  ")
    marks={}
    subjects=int(input("Enter number of subjects:"))
    
    for j in range(subjects):
        sub=input("Enter subject name: ")
        score=float(input("enter the score: "))
        marks[sub]=score
        
    student=Student(name,marks)
    students.append(student)    
    
print("\n---Student details---") 

for s in students:
    total=s.calculate_total()
    avg=s.calculate_avg()
    grade=s.get_grade()
    print(f"Name:{s.name}")
    print(f"Total marks:{total}")  
    print(f"Avarage marks: {avg}")  
    print(f"Grade: {grade}")                 
            
```
