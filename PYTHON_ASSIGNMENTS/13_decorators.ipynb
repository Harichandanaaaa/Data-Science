```python
def logger(func):
    def wrapper():
        print("Function execution started")
        func()
        print("Function execution ended")
    return wrapper

@logger
def proces_date():
    print("Processing data....")  
    
proces_date()                
```

```python
import time
def timer(func):
    def wrapper():
        start= time.time()
        func()
        end= time.time()
        print("Time:",end-start)
    return wrapper
@timer #Extending the functionality by passing another function as parameteer called slow
def slow():
    for i in range(1000000):
        pass
        
slow()  
```

```python
def auth(func):
    def wrapper():
        print("User authenticated")
        func()
    return wrapper

@auth
def home():
    print("Home page")
    

home()       
```

```python
def safe(func):
    def wrapper():
        try:
            func()
        except Exception as e:
            print("Error:",e)
    return wrapper

@safe
def divide():
    print(10/0)  
    
divide()             
```

```python
def  retry(func):
    def wrapper():
        for i in range(3):
            try:
                return func()
            except:
                print("Retrying...")
    return wrapper
@retry
def fetch_data():
    print(10/0)
    
fetch_data()                
```
