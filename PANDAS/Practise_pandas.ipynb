# Merging (JOIN)
Connecting related DataFrames based on keys.

```python
# Inner Join
merged_df = pd.merge(users_df, orders_df, on='user_id', how='inner')

# Left Join
merged_left = pd.merge(users_df, orders_df, left_on='id', right_on='user_id', how='left')
# how parameter accepts: 'left', 'right', 'outer', 'inner'.
```

```python
# Inner Join
user_df=pd.DataFrame({
    'user_id':[1,2,3],
    'Name':['Ravi','Pavan','Venkat']
})
order_df=pd.DataFrame({
    'user_id':[1,2,4],
    'Order':['Laptop','Mobile','Tablet']
})
merged_df=pd.merge(user_df, order_df, on='user_id', how='inner')
merged_df
```

```python
# Left Join
user_df=pd.DataFrame({
    'id':[1,2,3],
    'Name':['Ravi','Pavan','Venkat']
})
order_df=pd.DataFrame({
    'user_id':[1,2,4],
    'Order':['Laptop','Mobile','Tablet']
})
left_join=pd.merge(user_df,order_df, left_on='id',right_on='user_id',how='left')
left_join
```

# 9.Time Series Data Handling

Pandas excels at handling dates and time

```python
# Convert strings to datetime objects
df['Date']=pd.to_datetime(df['Date_String'])

# extract components
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
df['DayOfWeek']=df['Date'].dt.day_name()

# Setting datetime as index allows for powerful time slicing and resampling
df.set_index('Date', inplace=True)
df.loc['2023-01'] #Selects all rows in Jan 2023

# Resampling (e.g., converting daily data to monthly sum)
monthly_revenue=df['Revenue'].resample('M').sum()
```

```python
df=pd.DataFrame({
    "Date_String": ["2023-01-01", "2023-01-15", "2023-02-10", "2023-02-20"],
    "Revenue": [100, 200, 150, 250]
})
df['Date']=pd.to_datetime(df['Date_String'])#it creates another column name Date in dattime formate
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
df['DayOfWeek']=df['Date'].dt.day_name()
df.set_index('Date',inplace=True)
df.loc['2023-01']
monthly_revenue=df['Revenue'].resample('ME').sum()
print(monthly_revenue)
print(df)
```
