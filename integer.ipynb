```python
##print("\nQuestion 6: Check if 97 is a prime number")
num=97
count=0
for i in range(1,num+1):
  if count % i==0:
    count +=1
if count ==2:
  print("prime number:")
else:
  print("not a prime number:")
```

```python
##print("\nQuestion 10: Calculate the sum of first 20 odd numbers")
odd_num= 0
count = 0
num = 1

while count < 20:
    odd_num = odd_num + num
    num = num + 2
    count = count + 1

print(f"Sum of first 20 odd numbers is:{odd_num}")
```
