```python
# these are input columns
X =df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
X.head()
```

```python
X=df.drop("Price", axis=1)
```

```python
X.head()
#These Columns are independent Variables
```

```python
# Output Or Dependent or Traget variable
Y=df["Price"]
Y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X, Y, test_size= 0.2, random_state= 4)
```

```python
X_test.shape
```

```python
## Importinng model from the library.. (youll import all machine learning models from scikit-learn library)
# pip install scikit-learn  (run this code in your terminal to install scikit learn where youll get your ml models)
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
## for training youll have to give lm two parts which is X_train, y_train 
## lm.fit() this method is used to train your ml model
lm.fit(X_train,Y_train)
```

```python
print(lm.intercept_)
```

```python
## slops.. (Coefficient)
coeff_df=pd.DataFrame(lm.coef_,X.columns,columns=["Coefficient"])
```

```python
coeff_df
```

```python
## giving X_test to the model to make predictions.. 
# using trianig data to train the ml model .. liearn reg(X_train, y_train)
# after training model we will make preedictions on unseen data from the model 
prediction=lm.predict(X_test)
```

```python
prediction
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_test,prediction)
```

```python
sns.displot((Y_test-prediction),bins=50)
```

```python
## Regression Evaluation metrics..
from sklearn import metrics
print("Mean Absolute eroor",metrics.mean_absolute_error(Y_test,prediction))
print("Mean Squared eroor",metrics.mean_squared_error(Y_test,prediction))
print("Root mean Absolute eroor",metrics.root_mean_squared_error(Y_test,prediction))
```
